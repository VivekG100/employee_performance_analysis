```python
y = data["Performance_Score"]
x = data[["Years_At_Company","Monthly_Salary", "Overtime_Hours", "Promotions", "Employee_Satisfaction_Score"]]
```

```python
x.columns
```

```python
scaler= StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x,y,test_size= 0.2)
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_train
```

```python

import joblib
joblib.dump(scaler,"scaler.pkl")
```

```python
x_test= scaler.fit_transform(x_test)
```

```python
x_test
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
def modelperformance(predictions):
    print("accuracy score on model is {}" .format(accuracy_score(y_test, predictions)))
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(x_train, y_train)
```

```python
log_model.predict(x_test)
```

```python
log_predictions = log_model.predict(x_test)
```

```python
modelperformance(log_predictions)
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors": [3,5,7,9,11],"weights":["uniform","distance"]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gridkn= GridSearchCV(KNeighborsClassifier(), param_grid, cv = 5)
gridkn.fit(x_train, y_train)
```

```python
gridkn.best_params_
```

```python
predictions = gridkn.predict(x_test)
```

```python
modelperformance(predictions)
```

```python
from sklearn.svm import SVC
svm=SVC()
```

```python
param_grid={
    "C" : [0.01, 0.1, 0.5, 1],
    "kernel" : ["linear","rbf","poly"]
}
```

```python
gridsvc = GridSearchCV(svm, param_grid, cv=5)
```

```python
gridsvc.fit(x_train, y_train)
```

```python
y_pred = gridsvc.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
joblib.dump(gridsvc, "model.pkl")
```
